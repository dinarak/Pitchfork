### Trying to download all the info into one csv file gave me a timeout error 'ReadTimeout: HTTPConnectionPool(host='www.billboard.com', port=80): Read timed out. (read timeout=25)'.
### I am going to try to read in six month intervals.

```python
def datetimeIterator(from_date=None, to_date=None, delta=datetime.timedelta(minutes=1)):
    while to_date is None or from_date <= to_date:
        yield from_date
        from_date = from_date + delta
    return

for start_date in datetimeIterator(datetime.datetime(1998,1,3,0,0), datetime.datetime.today(), datetime.timedelta(weeks = 26)):
    
    chart_6mo = defaultdict(list)
    weekof = start_date

    while (weekof < start_date + datetime.timedelta(weeks = 26) and weekof < datetime.datetime.today()):

        # chart_array_thatweek = []
        weekof_str = weekof.strftime('%Y-%m-%d')
        chart  = billboard.ChartData('top-album-sales', date = weekof_str)

        for entry in chart:
            chart_6mo['date'].append(weekof_str)
            chart_6mo['title'].append(entry.title)
            chart_6mo['artist'].append(entry.artist)
            chart_6mo['peakPos'].append(entry.peakPos)
            chart_6mo['lastPos'].append(entry.lastPos)
            chart_6mo['weeks'].append(entry.weeks)
            chart_6mo['rank'].append(entry.rank)

        weekof += datetime.timedelta(weeks = 1)
        
    
    df = pd.DataFrame(chart_6mo)
    csv_name = 'billboard_st_at_'+ start_date.strftime('%Y-%m-%d')+'.csv'
    df.to_csv(csv_name)
```

### Concatenate all our Billboard files into one 

```python
import glob
import re
import pandas as  pd

files = glob.glob('billboard_st_at*.csv')

df_list = []
for filename in sorted(files):
    df_list.append(pd.read_csv(filename, encoding="ISO-8859-1"))
    
full_df = pd.concat(df_list)

full_df.to_csv('billboard_albums.csv')
```

### Aggregate data for each album: first week it appeared in the chart, its peak position ever, max number of weeks on the chart, and its average rank in the chart

```python
import numpy as np

grouped = albums.groupby(['artist', 'title'])
albums_clnd = grouped.agg({'date' : 'min',
            'peakPos': 'max', 
            'weeks': 'max', 
            'rank': 'mean'})
albums_clnd.head()
```
